importing and reading files

```python
import pandas as pd
url = "https://raw.githubusercontent.com/KaranSinghS/IBM_internship/main/SampleSuperstore.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.drop(columns="Country")
df.drop(columns="Postal Code")
```

```python
df.describe()
```

```python
df.info()
```

```python
df.groupby('Region')['Sales'].sum().plot.bar()
```

```python
df.groupby("Region")["Profit"].sum().plot.bar()
```

```python
df.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
df.groupby("Segment")["Profit"].sum().plot.bar()
```

```python
df.groupby("Category")["Profit"].sum().plot.bar()
```

```python
df.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
df.groupby("State")["Sales"].sum().plot.bar()
```

```python
df.groupby("State")["Profit"].sum().plot.bar()
```

```python
df.groupby("Category")["Profit","Sales"].sum().plot.bar()
```

```python
df.groupby("Segment")["Profit","Sales"].sum().plot.bar()
```

```python
df.groupby("Region")["Profit","Sales"].sum().plot.bar()
```
